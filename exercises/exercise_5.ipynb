Test the PCA class in a jupyter notebook using the iris.csv dataset (classificação).

```python
import numpy as np
import matplotlib.pyplot as plt
from si.io.csv_file import read_csv
from si.decomposition.pca import PCA
```

```python
path = "../datasets/iris/iris.csv"
dataset = read_csv(filename=path, features=True, label=True)
```

```python
# Map string labels to numerical values
label_mapping = {label: idx for idx, label in enumerate(set(dataset.y))}
numeric_labels = np.array([label_mapping[label] for label in dataset.y])
```

### 1D PCA (n_components=1)

```python
pca = PCA(dataset=dataset, n_components=1)
pca = pca.fit(dataset)
X_reduced = pca.transform(dataset)
print(pca.explained_variance)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], np.zeros_like(X_reduced[:, 0]), c=numeric_labels, cmap="viridis", edgecolor="k")

plt.xlabel("Principal Component 1")
plt.ylabel("Zero (arbitrary)")
plt.title("PCA Visualization (1D)")

plt.colorbar(scatter, label="Classes")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], np.arange(len(X_reduced)), c=numeric_labels, cmap="viridis", edgecolor="k")

plt.xlabel("Principal Component 1")
plt.ylabel("Sample Index")
plt.title("PCA Visualization (1D)")

plt.colorbar(scatter, label="Classes")
plt.show()
```

### 2D PCA (n_components=2)

```python
pca = PCA(dataset=dataset, n_components=2)
pca = pca.fit(dataset)
X_reduced = pca.transform(dataset)
print(pca.explained_variance)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=numeric_labels, cmap="viridis", edgecolor="k")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Visualization")

plt.colorbar(scatter, label="Classes")

plt.show()
```

### 3D PCA (n_components=3)

```python
pca = PCA(dataset=dataset, n_components=3)
pca = pca.fit(dataset)
X_reduced = pca.transform(dataset)
print(pca.explained_variance)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2],
    c=numeric_labels, cmap="viridis", edgecolor="k"
)

ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("3D PCA Visualization")

fig.colorbar(scatter, ax=ax, label="Classes")

plt.show()
```
